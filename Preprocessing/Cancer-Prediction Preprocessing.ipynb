```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_550 = pd.read_csv('Sample size of even 550.csv')
df_942 = pd.read_csv('Sample size of even 942.csv')
df_1636 = pd.read_csv('Sample size of even 1636.csv')
df_1951 = pd.read_csv('Sample size of even 1951.csv')
```

```python
#we're combining the csvs and removing the nan data in the result column
combined_df = pd.concat([df_550,df_942,df_1636,df_1951],ignore_index=True)
```

```python
df = combined_df[combined_df['Result'].notna()]
```

```python
df.shape
```

```python
df.drop(df.columns[0], axis=1,inplace=True)
df.drop(['Ethnicity','Gender'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#converting the result column from text to float
df['Result'] = df['Result'].str.replace('%', '').astype(float)
```

```python
df.describe()
```

```python
#combining the smoking age and quit age into one column and remove Smoking Age and Quit Age
def calculate_years_smoking(row):
    
    if pd.isna(row['Quit_Age']):
        return row['Smoking_Age']
    else:
        return row['Quit_Age'] - row['Smoking_Age']

df['Years Smoking'] = df.apply(calculate_years_smoking, axis=1)
```

```python
df.drop(['Smoking_Age','Quit_Age','Cigarettes'],axis=1,inplace=True)
```

```python
df['Years Smoking']= df['Years Smoking'].fillna(0)
```

```python
df.head()
```

```python
#label encoding columns: Meds_one, Meds_two, Family History, Family Numbers, Colonoscopy

df['Meds_one'] = df['Meds_one'].map({'Yes':1,'No':0})
```

```python
df['Meds_Two'] = df['Meds_Two'].map({'Yes':1,'No':0})
```

```python
df['Colonoscopy'] = df['Colonoscopy'].map({'Yes':1,'No':0})
```

```python
df['Family History'] = df['Family History'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
df['Family Numbers'] = df['Family Numbers'].map({'NaN':0,'1':1,'2 or more':2})
```

```python
df['Family Numbers']= df['Family Numbers'].fillna(0)
```

```python
df.head()
```

```python
df.drop(['Currently Smoking','No of Cigarettes'],axis=1)
```

```python
df.drop(['Vig Activity','Vig Hours','Currently Smoking','No of Cigarettes'],axis=1,inplace=True)
```

```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
diet_servings_map = {
    'None': 0,
    'Less than 1 serving/week': 1,
    '1-2 servings/week': 2,
    '3-4 servings/week': 3,
    '5-6 servings/week': 4,
    '7-10 servings/week': 5,
    'More than 10 servings/week': 6
}

num_servings_map = {
    ' cup or less': 1,
    'Between  cup - 1 cups': 2,
    '1 cups - 3 cups': 3,
    '3 cups - 5 cups': 4,
    'More than 5 cups': 5
}
mod_hours = {'Up to 1 hour/week':0,
                'Between 1 - 2 hours/week':1,
                '2 - 3 hours/week':2,
                '3 - 4 hours/week':3,
                'More than 4 hours/week':4}


```

```python
df_clean['Diet Servings'] = df_clean['Diet Servings'].map(diet_servings_map)
```

```python
df_clean['Number of Servings'] = df_clean['Number of Servings'].map(num_servings_map)
```

```python
df_clean['Mod_Hours'] = df_clean['Mod_Hours'].map(mod_hours)
```

```python
df_clean['Number of Servings'] = df_clean['Number of Servings'].fillna(0)
```

```python
df_clean.head()
```

```python
#creating the Colonoscopy / Polyp interaction
df_clean['Polyp'] = df_clean['Polyp'].map({'Yes': 1, 'No': 0})
```

```python
df_clean['Polyp'] = df_clean['Polyp'].fillna(-1)
```

```python
df_clean['Colonoscopy_Polyps_Interaction'] = df_clean['Colonoscopy'] * df_clean['Polyp']
```

```python
df_clean.head()
```

```python
df_clean.drop(['Colonoscopy','Polyp'],axis=1,inplace=True)
```

```python
df_clean.head()
```

```python
df_clean.drop(['Meds_one','Meds_Two','Family History'],axis=1,inplace=True)
```

```python
df_clean.to_csv('Cancer Preprocessed February Part Deux.csv', index=False)  
```
